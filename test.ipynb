Given $N$ neurons, record full weight matrix $W = (N \times N)$ s.t. $W(i,j) = - W(j, i)$ and $W(i,i) = 0$ for $i > j$. 
Activation path is those with positive connection weights. Operation on input $f(X) = X - W \cdot X$, to minic the 'resistance' nature of neural propagation.
